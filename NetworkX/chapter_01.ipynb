```python
# Configure plotting in Jupyter
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams.update({
    'figure.figsize': (7.5, 7.5),
    'axes.spines.right': False,
    'axes.spines.left': False,
    'axes.spines.top': False,
    'axes.spines.bottom': False})
```

```python
# imports NetworkX package with nx alias
import networkx as nx

# Create a Graph object representing an undirected network
G = nx.Graph()
# Add nodes and edges to the network using letters as node IDs. Each node has to have a unique ID- 
# can be a number, string, tuple, any Python object that has __hash__() method defined
G.add_nodes_from(['A', 'B', 'C'])
G.add_edge('A', 'B')
G.add_edges_from([('B', 'C'), ('A', 'C')])

# Draw a simple visualization of this network. 
# figure() function is used to create a 7.5 by 7.5 inch figure
# draw_networkx() uses G to produce a visualization 
plt.figure(figsize = (7.5, 7.5))
nx.draw_networkx(G)
plt.show() # can be omitted when using Jupyter Lab
```

```python
# Sets random seeds (configures randomized algorithm to produce the same output each time)
import random
from numpy import random as nprand
seed = hash("Network Science in Python") % 2**32 
nprand.seed(seed)
random.seed(seed)
```
